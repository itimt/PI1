**Configurações Iniciais**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

!wget -O breja.jpg https://raw.githubusercontent.com/itimt/PI1/main/Breja.jpeg
```

***Exercício 1: Carregamento e Correção de Canais***

```python
img_bgr = cv2.imread('breja.jpg')

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Exercício 1: Imagem Original (RGB)")
plt.axis('off')
plt.show()
```

***Exercício 2: Atributos da Imagem***

```python
print(f"Shape (Dimensões): {img_rgb.shape}")
print(f"Size (Total de Pixels): {img_rgb.size}")
print(f"Dtype (Tipo de Dado): {img_rgb.dtype}")
```

***Exercício 3: Salvar em Formato Diferente***

```python
cv2.imwrite('breja_copia.png', img_bgr) # Usamos img_bgr pois cv2.imwrite espera BGR
print("Cópia salva como 'breja_copia.png'")
```

***Exercício 4: Acessar Pixel Central***

```python
altura, largura, _ = img_rgb.shape
centro_y, centro_x = altura // 2, largura // 2

pixel_central = img_rgb[centro_y, centro_x]
r, g, b = pixel_central

print(f"Coordenadas centrais: Y={centro_y}, X={centro_x}")
print(f"Valores no centro - R: {r}, G: {g}, B: {b}")
```

***Exercício 5: Modificar Região (Cor Sólida)***

```python
img_modificada = img_rgb.copy()

img_modificada[0:100, 0:100] = [0, 0, 255]

plt.imshow(img_modificada)
plt.title("Exercício 5: Quadrado Azul no Canto")
plt.axis('off')
plt.show()
```

***Exercício 6: Recorte (ROI - Region of Interest)***

```python
recorte = img_rgb[300:600, 300:700]

plt.imshow(recorte)
plt.title("Exercício 6: Recorte (ROI)")
plt.axis('off')
plt.show()
```

***Exercício 7: Separar Canais B, G, R***

```python
b, g, r = cv2.split(img_bgr)
print("Canais separados com sucesso!")
```

***Exercício 8: Exibir Canal Verde (G) em Cinza***

```python
plt.imshow(g, cmap='gray')
plt.title("Exercício 8: Canal Verde (G)")
plt.axis('off')
plt.show()
```

***Exercício 9: Merge com Troca de Canais (Efeito Visual)***

```python
# Realizando o merge trocando as posições: R no lugar de B
img_merge_efeito = cv2.merge([r, g, b])

plt.imshow(img_merge_efeito)
plt.title("Exercício 9: Merge com Canais Trocados")
plt.axis('off')
plt.show()
```

```python
# Realizando o merge trocando as posições: B no lugar de R
img_merge_efeito = cv2.merge([b, g, r])

plt.imshow(img_merge_efeito)
plt.title("Exercício 9: Merge com Canais Trocados")
plt.axis('off')
plt.show()
```

***Exercício 10: Converter para Tons de Cinza***

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.title("Exercício 10: Imagem em Tons de Cinza")
plt.axis('off')
plt.show()
```

***Exercício 11: Espaço HSV e Canal de Saturação (S)***

```python
img_hsv = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2HSV)

h, s, v = cv2.split(img_hsv)

plt.imshow(s, cmap='gray')
plt.title("Exercício 11: Canal de Saturação (S)")
plt.axis('off')
plt.show()
```

***Exercício 12: Aumento de Brilho com Tratamento de Transpordo***

```python
valor_brilho = np.ones(img_rgb.shape, dtype="uint8") * 50

img_brilhante = cv2.add(img_rgb, valor_brilho)

plt.imshow(img_brilhante)
plt.title("Exercício 12: Brilho Aumentado (+50)")
plt.axis('off')
plt.show()
```

***Exercício 13: Mesclagem de Duas Fotos***

```python
!wget -O aviao.jpeg https://raw.githubusercontent.com/itimt/PI1/main/aviao.jpeg

img2_bgr = cv2.imread('aviao.jpeg')

larg, alt = 800, 600
img1_res = cv2.resize(img_rgb, (larg, alt))
img2_res = cv2.resize(cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2RGB), (larg, alt))

img_blend = cv2.addWeighted(img1_res, 0.5, img2_res, 0.5, 0)

plt.imshow(img_blend)
plt.title("Exercício 13: Mesclagem (Blend)")
plt.axis('off')
plt.show()
```

***Exercício 14: Negativo Fotográfico***

```python
img_negativa = 255 - img_gray

plt.imshow(img_negativa, cmap='gray')
plt.title("Exercício 14: Negativo Fotográfico")
plt.axis('off')
plt.show()
```

***Exercício 15: Máscara Binária Circular***

```python
mask = np.zeros(img_gray.shape, dtype="uint8")

centro = (img_gray.shape[1] // 2, img_gray.shape[0] // 2)
cv2.circle(mask, centro, 300, 255, -1) # Raio 300, cor branca, preenchido (-1)

img_focada = cv2.bitwise_and(img_gray, img_gray, mask=mask)

plt.imshow(img_focada, cmap='gray')
plt.title("Exercício 15: Máscara Circular")
plt.axis('off')
plt.show()
```

***Exercício 16: Redução de Tamanho (30%)***

```python
# Calculando 30% do tamanho original
largura_n = int(img_rgb.shape[1] * 0.3)
altura_n = int(img_rgb.shape[0] * 0.3)
nova_dim = (largura_n, altura_n)

img_reduzida = cv2.resize(img_rgb, nova_dim, interpolation=cv2.INTER_AREA)

print(f"Shape original: {img_rgb.shape}")
print(f"Novo shape (30%): {img_reduzida.shape}")
plt.imshow(img_reduzida)
plt.axis('off')
plt.show()
```

***Exercício 17: Valores de Intensidade do Histograma***

```python
hist_valores = cv2.calcHist([img_gray], [0], None, [256], [0, 256])

for i in range(10):
    print(f"Intensidade {i}: {hist_valores[i][0]}")
```

***Exercício 18: Plote do Histograma Colorido***

```python
cores = ('r', 'g', 'b')
plt.figure()
plt.title("Exercício 18: Histograma por Canais")
for i, col in enumerate(cores):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0, 256])
plt.show()
```

***Exercício 19: Equalização de Histograma (Contraste)***

```python
img_equalizada = cv2.equalizeHist(img_gray)

plt.subplot(1, 2, 1), plt.imshow(img_gray, cmap='gray'), plt.title("Original")
plt.subplot(1, 2, 2), plt.imshow(img_equalizada, cmap='gray'), plt.title("Equalizada")
plt.show()
```

***Exercício 20: Limiarização (Threshold) Binária***

```python
_, img_threshold = cv2.threshold(img_gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(img_threshold, cmap='gray')
plt.title("Exercício 20: Threshold Binário")
plt.axis('off')
plt.show()
```
